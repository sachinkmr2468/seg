```python
import requests
import pandas as pd

url1 = 'https://api.nobelprize.org/v1/laureate.json'
url = requests.get(url1).json()
df = pd.DataFrame(url['laureates'])
df
```

```python
df1 = df[['id','firstname','surname','born','prizes','gender','bornCountryCode']]
df1['prizes'] = df['prizes'].apply(lambda x:x[0]['year'])
df1['category'] = df['prizes'].apply(lambda x:x[0]['category'])
df2 = df1.drop_duplicates(subset='prizes')
#df3 = df2.drop_duplicates(subset='category')
#df2.to_csv(r'C:\Users\sachikum\Documents\ML interview\segregate4.csv')
df2
```

```python
lst = df.bornCountryCode.to_list()
lst
```

```python
import pycountry
def get_country(n):
    country = pycountry.countries.get(alpha_3 = n)
    if country:
        return country.name
    else:
        return n
country_name = [get_country(name) for name in lst]
df["bornCountryCode"] = country_name
df.head()
```
